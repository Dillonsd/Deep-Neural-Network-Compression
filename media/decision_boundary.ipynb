```python
import matplotlib.pyplot as plt
import numpy as np

weight = [1, 1]
bias = 2

sample_class_a = np.random.normal(loc=[1.5, 1.5], scale=1, size=(100, 2))
sample_class_b = np.random.normal(loc=[-3.5, -3.5], scale=1, size=(100, 2))

# Plot the decision boundary
x = np.linspace(-5, 5, 100)
y = -weight[0] / weight[1] * x - bias / weight[1]
plt.plot(x, y, 'k-')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.quiver(0, -2, weight[0], weight[1], scale=10, color='red')
plt.grid(True)
plt.plot(sample_class_a[:, 0], sample_class_a[:, 1], 'bo')
plt.plot(sample_class_b[:, 0], sample_class_b[:, 1], 'ro')
plt.show()
```
