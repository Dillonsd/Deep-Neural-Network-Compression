```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y_relu = np.maximum(x, 0)
y_sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y_relu)
plt.title('ReLU')
plt.subplot(1, 2, 2)
plt.plot(x, y_sigmoid)
plt.title('Sigmoid')
plt.show()
```
